### Declare Library

```python
# web scraping

import requests
import matplotlib.pyplot as plt 
from bs4 import BeautifulSoup as bs
import pandas as pd

import numpy as np

```

### Web 1 (Negara dan Populasi)

```python
web1 = requests.get ('https://www.worldometers.info/geography/alphabetical-list-of-countries/')
web1
```

##### Nama-nama negara

```python
# declare the data from web and find the table

data_soup1 = bs (web1.text, 'lxml')
table = data_soup1.find ('table')
```

##### Populasi

```python
populasi = table.find_all ('a')
populasi
```

### Web 2 (Militer Negara)

```python
# Tes perizinan web (404 = ga boleh, 200 = boleh) 

web2 = requests.get ('https://www.globalfirepower.com/countries-listing.php')
web2
```

```python
data_soup2 = bs (web2.text, 'lxml')
nama_negara = data_soup2.find_all ('div', 'longFormName')

for i in nama_negara:
    value = i.find ('span', 'textWhite textLarge textShadow').text
    print (value)
```

### Web 3 (List Negara Berdasarkan Alpahabet)

```python
web3 = requests.get ('https://www.britannica.com/topic/list-of-countries-1993160')
web3
```

```python
data3 = bs (web3.text, 'lxml')
nama_negara_alphabetical = data3.find_all ('a', 'md-crosslink')
nama_negara_alphabetical
```

```python
unwanted_words = ["geography", "government"]
target = [Nama_negara_alphabetical.text for Nama_negara_alphabetical in nama_negara_alphabetical if Nama_negara_alphabetical.text not in unwanted_words]
target
```
